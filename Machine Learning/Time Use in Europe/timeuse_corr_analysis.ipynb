### Retrieving frame with Top correlated variables

```python
counts = []

for col in timeuse.columns[4:]:
    
    col_corr = time_men.corr()[[col]]
    col_cnt70 = col_corr[numpy.abs(col_corr[col])>0.85][col].count()
    
    counts.append([col, col_cnt70])
    
corr_rank = pandas.DataFrame(counts, columns=['Activity', 'N° of corr.']).sort_values('N° of corr.', ascending=False)
    
```
