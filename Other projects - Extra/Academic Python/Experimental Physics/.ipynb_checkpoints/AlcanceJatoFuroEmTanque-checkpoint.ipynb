```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

### Definindo os Pares Ordenados

```python
#D = [26.58, 24.3, 20.7, 16.58, 11.97]
```

```python
D = [26.62, 23.5, 20.47, 16.99, 12.12]
H = [10, 8, 6, 4, 2]
D_2 = [i**2 for i in D]

X = np.array(H).reshape(-1, 1)
Y = D_2
```

```python
X
```

```python
Y
```

### Ajustando a Regressão

```python
LR = lr().fit(X, Y)
a = LR.coef_
b = LR.intercept_

a, a/4, b
```

```python
#x2 = 4Hy; a = 4H
```

```python
y = a/4; y  # (resposta)
```

### Visualizando a Regressão

```python
import seaborn as sns
```

```python
sns.set()
```

```python
xx = np.linspace(0,10)
yy = a * xx + b

fig = plt.figure(figsize=(5,3))
plt.scatter(X, Y)
plt.plot(xx, yy)

plt.ylim(-30); plt.xlim(-.4)
plt.title('Regressão Linear Simples')
plt.xlabel("X ~ Altura coluna d'água (m)")
plt.ylabel("Y ~ Distância jato d'água^2 (m2)")
# plt.grid()

plt.savefig('reg.jpg', bbox_inches='tight')
plt.show()
```

```python
xx = np.linspace(0,10)
yy = a * xx + b

fig = plt.figure(figsize=(5,3))
plt.scatter(X, Y)
plt.plot(xx, yy)

plt.ylim(-30); plt.xlim(-.4)
plt.title('Regressão Linear Simples')
plt.xlabel("X ~ Altura coluna d'água (m)")
plt.ylabel("Y ~ Distância jato d'água^2 (m2)")
plt.grid()

plt.savefig('reg.jpg', bbox_inches='tight')
plt.show()
```
