# Discursiva

A frequência observada pelo motorista é menor que a frequência emitida pela ambulância, portanto os dois estão se afastando e Vobs > Vf. 

Sabe-se que o observador está se movendo para longe da fonte. Portanto, o sinal da velocidade do observador (Vobs) na formula da frequência observada será negativo. Entretanto, Vf pode estar direcionada para o observador ou para longe dele, o que será determinado.

Para descobrir se a ambulância está indo na direção do carro ou para longe dele pode-se usar a variável Vf (velcidade da fonte) com sinal positivo na fórmula da frequência observada. Vf pode ser então isolada e seu valor determinado. Caso seu valor retorne positivo, a fonte estará indo na direção oposta àquela do carro. Caso contrário, se o sinal encontrado para Vf for negativo a fonte estará indo em direção ao carro. Esse raciocínio pode ser confirmado pelo fato de que sinal positivo para Vf aumenta o valor do denominador, o que gera uma diminuição da razão da frequência observada e da emitida. Logo, o sinal de Vf positivo gera uma frequência observada menor em relação a emitida do que o sinal negativo para Vf e, portanto, considerando o efeito doppler, Vf positivo indicará que a fonte está se afastando do observador e vice-versa.  Assim,

f' = fo (Vonda - Vobs) / (Vonda + Vf)

Onde,

f' é a frequência observada em Hz,
fo é a frequência emitida pela fonte em Hz,
Vonda é a velocidade da onda sonora no ar em m/s,
Vobs é a velocidade do observador em m/s (negativa pois o observador vai na direção oposta a da fonte),
Vf é a velocidade da fonte em m/s (positiva para que seu sinal seja determinado).

A velocidade do carro (observador) convertida para metros por segundo:

Vobs = (72 Km/h) / (3,6 h/s) = 20 m/s

As demais variáveis:

Vonda = 340 m/s
fo = 850 Hz
f' = 780 Hz

Isolando Vf:

f' = fo (Vonda - Vobs) / (Vonda + Vf)

Vonda + Vf = (Vonda - Vobs) fo/f'

Vf = (Vonda - Vobs).(fo/f') - Vonda

Substituindo os valores:

Vf = (850 Hz / 780 Hz) * (340 m/s - 20 m/s) - 340 m/s

Vf = 8,7 m/s

Como o sinal encontrado é positivo, a ambulância está indo no sentindo oposto ao do carro.
