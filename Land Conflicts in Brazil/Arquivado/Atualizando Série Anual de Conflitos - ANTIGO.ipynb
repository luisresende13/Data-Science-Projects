# Build sorted pivot of the conflicts dataset

## Sort by city and year

```python
from copy import deepcopy

def multiclass_column_pivot(df, key, sep=', '):
    multiclass_list = df[key].apply(lambda value: value.split(sep)).tolist()
    df = df.copy(); n_rows = len(df)
    index = df.index.tolist()
    dup_rows, new_rows = [], []
    for i, idx in enumerate(index):
        if i%20 == 0: 
            co(True)
            print(f'rows spreaded: {i+1}/{n_rows}')
        values = multiclass_list[i]
        if len(values) > 1:
            dup_rows.append(idx)
        row = df.loc[idx].copy()
        for value in values:
            row[key] = value
            new_rows.append(row.copy())
    return pd.concat(new_rows, axis=1).T.reset_index(drop=True)

key = 'Município'
df = land

df_pivot = multiclass_column_pivot(df, key, sep=', ')

df_pivot['Ano'] = pd.DatetimeIndex(df_pivot['Data']).year

sort_cols = ['Município', 'Ano', 'Data']
df_sort = pd.concat([df_pivot[sort_cols], df_pivot.drop(sort_cols, axis=1)], axis=1).sort_values(sort_cols)
df_sort.head()
```

## Data cleaning: Land dataset pivot 2011-2021

```python
# missing = list(set(cpt['MUN'].tolist()).difference(count.index.tolist()))
# print('Cities missing in primary dataset:\n\n', missing); print('\n')

missing = list(set(df_sort['Município'].tolist()).difference(cpt['MUN'].tolist()))
print('Cities missing in secondary dataset:\n\n', missing); print('\n')

# City renaming map
city_name_map = {
    'Diamante;Oeste': "Diamante D'Oeste",
    'Glória;Oeste': "Glória D'Oeste",
    'Pimenteiras': "Pimenteiras do Oeste",
    'Alta Floresta;Oeste': "Alta Floresta D'Oeste",
    'Alvorada;Oeste': "Alvorada D'Oeste",
    'Conquista;Oeste': "Conquista D'Oeste",
    'Nova Brasilândia;Oeste': "Nova Brasilândia D'Oeste",
    'Sant&#39;Ana do Livramento': "Sant'Ana do Livramento",    
}

# RENAME CITIES TO MATCH PREVIOUS STANDARD. 
df_sort['Município'] = df_sort['Município'].apply(lambda value: value if value not in city_name_map else city_name_map[value])

# Show that there are no cities missing in secondary dataset anymore.
missing = list(set(df_sort['Município'].tolist()).difference(cpt['MUN'].tolist()))
print('Cities missing after city name formatting:\n\n', missing)
```

## Merge conflics' history dataset with cities' shapes

#### Reload update time series

```python
import pandas as pd

conflicts_history = pd.read_csv('Dados/Resultados/ter-city-frontier-2000-2021.csv')

print(conflicts_history.shape)
```

#### Read cities geometries dataset

```python
import pandas as pd

cities = pd.read_csv('Dados/Resultados/BR_Municipios_2021.csv')
```

#### Add id field based on city and state

```python
conflicts_history['id'] = conflicts_history['MUN'].str.strip().str.lower() + ' - ' + conflicts_history['UF'].str.strip().str.lower()
cities['id'] = cities['NM_MUN'].str.strip().str.lower() + ' - ' + cities['SIGLA'].str.strip().str.lower()
```

#### Check compatibility

```python
print(conflicts_history.shape)
print(cities.shape)
print(conflicts_history['UF'].str.strip().isin(cities['SIGLA']).mean())
print(conflicts_history['id'].isin(cities['id']).mean())
print(len(conflicts_history['id'].unique()), len(conflicts_history))
print(len(cities['id'].unique()), len(cities))
```

#### Merge tables

```python
cities_conflicts = cities.join(conflicts_history.set_index('id'), on='id', how='inner')

# Drop commom columns
cities_conflicts.drop(['MUN', 'UF'], axis=1, inplace=True)

# display(cities_conflicts.head())
cities_conflicts.shape
```

### Include frontier flag

```python
state_abbr_dict = {
    'Acre': 'AC',
    'Alagoas': 'AL',
    'Amapá': 'AP',
    'Amazonas': 'AM',
    'Bahia': 'BA',
    'Ceará': 'CE',
    'Distrito Federal': 'DF',
    'Espírito Santo': 'ES',
    'Goiás': 'GO',
    'Maranhão': 'MA',
    'Mato Grosso': 'MT',
    'Mato Grosso do Sul': 'MS',
    'Minas Gerais': 'MG',
    'Pará': 'PA',
    'Paraíba': 'PB',
    'Paraná': 'PR',
    'Pernambuco': 'PE',
    'Piauí': 'PI',
    'Rio de Janeiro': 'RJ',
    'Rio Grande do Norte': 'RN',
    'Rio Grande do Sul': 'RS',
    'Rondônia': 'RO',
    'Roraima': 'RR',
    'Santa Catarina': 'SC',
    'São Paulo': 'SP',
    'Sergipe': 'SE',
    'Tocantins': 'TO'
}


# Create unique id reference

frontier_cities['UF'] = frontier_cities['Estado'].map(state_abbr_dict)
frontier_cities['id'] = frontier_cities['Município'].str.strip().str.lower() + ' - ' + frontier_cities['UF'].str.strip().str.lower()
cities_conflicts['id'] = cities_conflicts['NM_MUN'].str.strip().str.lower() + ' - ' + cities_conflicts['SIGLA'].str.strip().str.lower()
```

##### Check compatibility

```python
print('Ids de fronteira na tabela (%):', frontier_cities['id'].isin(cities_conflicts['id']).mean() * 100)
print('Cidades de fronteira na tabela (%):', frontier_cities['Município'].isin(cities_conflicts['NM_MUN']).mean() * 100)
print('Ids da tabela em fronteira (%):', cities_conflicts['id'].isin(frontier_cities['id']).mean() * 100)
print('Cidades na tabela em fronteira (%):', cities_conflicts['NM_MUN'].isin(frontier_cities['Município']).mean() * 100)
```

##### Add frontier flag

```python
cities_conflicts['Fronteira'] = cities_conflicts['id'].isin(frontier_cities['id']).astype(int)
```

#### Save as csv

```python
# Replace 'output.csv' with the desired output csv file path
csv_output_path = 'Dados/Resultados/BR_Municipios_2021_Conflitos.csv'

cities_conflicts.to_csv(csv_output_path)

cities_conflicts.head()
```

## Exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Reload update time series

```python
cpt_up_inner = pd.read_csv('Dados/Resultados/ter-city-frontier-2000-2021.csv').drop('id', 1)
print(cpt_up_inner.shape)

# standard time serie format
ts = cpt_up_inner.set_index('MUN').T.iloc[5:]

ts.index = list(map(lambda year: int(year[4:]), ts.index))
```

#### Filter cities in national frontier

```python
in_frontier = frontier_cities['Município']

ts_frontier = ts.iloc[:, ts.columns.isin(in_frontier)]
```

#### Plot time serie comparison

```python
fig, axs = plt.subplots(2, 1, figsize=(7, 9), tight_layout=True)

ts.plot(
    legend=[],
    title='Conflitos por terra em municípios próximos à fronteira - série anual',
    xlabel='Ano', ylabel='Conflitos', ax=axs[0]
)

ts_frontier.plot(
    legend=[],
    title='Conflitos por terra em municípios na fronteira - série anual',
    xlabel='Ano', ylabel='Conflitos', ax=axs[1]
)

print(f'- Cidades: {ts.shape[1]}')
print(f'- Cidades de Fronteira: {ts_frontier.shape[1]} · {round(ts_frontier.shape[1] / ts.shape[1] * 100, 1)}%')

plt.show()
```
