#### Importing mongodb module

```python
import pymongo
import pandas as pd
import json
import pprint
```

## MongoDB Authentication

```python
password = 'admin'
database = 'MoodTrackerDatabase'

uri = f"mongodb+srv://admin:{password}@mood-tracker-cluster.f0b5r.mongodb.net/{database}?retryWrites=true&w=majority"
#uri = f"mongodb://admin:{password}@mood-tracker-cluster.f0b5r.mongodb.net/{database}?retryWrites=true&w=majority"
```

```python
client = pymongo.MongoClient(uri)
```

## Retrieving app users database

```python
db = client[database]
```

```python
users = db.Users
```

```python
users_list = []
for user in users.find():
    users_list.append(user)
```

---

## Cleaning Data

### Converting json database to pandas dataframe

```python
users_df = pd.DataFrame(users_list)
```

### Json dumping objects

```python
for entries in users_df['entries']:
    for entry in entries:
        entry['user_id'] = str(entry['user_id'])
        entry['_id'] = str(entry['user_id'])        
```

```python
users_df_json = users_df.copy()
```

```python
new_users_entries = []
for entries in users_df_json['entries']:
    new_users_entries.append(json.dumps(entries))

users_df_json['entries'] = new_users_entries
```

### Saving Clean Data

```python
users_df_json.to_csv('UsersDatabase.csv', index=False)
```
