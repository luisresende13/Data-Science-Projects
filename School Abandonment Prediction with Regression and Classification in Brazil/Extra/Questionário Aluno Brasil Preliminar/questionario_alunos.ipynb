#### Label Encoding Features

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labeled = ['ID_MUNICIPIO', 'ID_ESCOLA', 'ID_TURMA', 'ID_ALUNO', 'ID_DEPENDENCIA_ADM', 'ID_TURNO', 'ID_LOCALIZACAO', 'ID_CAPITAL', 'ID_REGIAO']

for col in labeled:
    
    ids[col] = ids[col].astype(str)
    ids[col] = LabelEncoder().fit(ids[col].unique()).transform(ids[col]) 
    
    
```
