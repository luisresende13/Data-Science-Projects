#### Importing Necessary Modules

```python
import pandas
import numpy
import matplotlib.pyplot as plt

```

### Loading Time Use in Europe Data

```python
timeuse = pandas.read_csv('/Users/luisr/Desktop/Repository/IBM_courses/Samples/timeuse_cleaned.csv')
timeuse.drop('Unnamed: 0', axis=1, inplace=True)
```

### Uncovering Most Correlated Domestic Activities

```python
dish_wash = timeuse.corr()[['Dish washing']].sort_values('Dish washing', ascending=False)

dish_wash.head(7)
```

### Sampling Target and Features

```python
sample = timeuse[['SEX', 'GEO/ACL00'] + dish_wash.index[:6].values.tolist()]
```

```python
sample.head(3)
```

### Pre-processing Sample Data

#### Label Encoding:

```python
from sklearn import preprocessing

sample['SEX'] = preprocessing.LabelEncoder().fit(['Males', 'Females']).transform(sample.loc[:,'SEX'])

sample['GEO/ACL00'] = preprocessing.LabelEncoder().fit(sample['GEO/ACL00'].unique()).transform(sample['GEO/ACL00'])
```

```python
sample.head()
```

### Defining Target and Features

```python
y=sample[['SEX']]
x=sample.drop('SEX', axis=1)
```

### Spliting Sample into Train and Test Sets

```python
from sklearn import model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.2, random_state=2)
```

### Fitting Tree Model

```python
from sklearn import tree

tree_obj = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4).fit(x_train, y_train)
```

### Predicting for Test and Train Sets

```python
y_hat_tst = tree_obj.predict(x_test)
#y_hat_trn = tree_obj.predict(x_train)
```

### Scoring Predictions

```python
from sklearn import metrics

scr_tst = metrics.accuracy_score(y_test, y_hat_tst)
scr_trn = metrics.accuracy_score(y_train, y_hat_trn)
```

### Visualizing Tree

```python
from io import StringIO
import pydotplus
from matplotlib import image
from sklearn import tree
```

```python
dot_data = StringIO()
feature_names = x.columns[:]
class_names = y.columns[0]

OUT = tree.export_graphviz(tree_obj,
                           out_file=dot_data,
                           feature_names=feature_names,
                           class_names=class_names,
                           filled=True,
                           special_characters=True,
                           rotate=False)

file_name = 'sex_domestic_tree.png'

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(file_name)

img = image.imread(file_name)
plt.figure(figsize=(100, 200))
plt.imshow(img, interpolation='nearest')
```
