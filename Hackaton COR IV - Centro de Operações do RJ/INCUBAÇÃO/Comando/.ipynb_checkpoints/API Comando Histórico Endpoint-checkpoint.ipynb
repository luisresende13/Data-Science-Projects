# *SISTEMA COMANDO'S* HISTORY THROUGH API ENDPOINT

### Comando API request functions

```python
# Python modules
import requests, json, time, pandas as pd, numpy as np, urllib
from matplotlib.path import Path as mpl_path

from datetime import datetime
from matplotlib.path import Path as mpl_path
from IPython.display import clear_output as co
from google.cloud import bigquery
from google.oauth2 import service_account

# google cloud auth
project_id = 'pluvia-360323'
google_credentials = '../../../../Apps/Python/bolsao-api/credentials/pluvia-360323-eae2907a9c98.json'
google_credentials = service_account.Credentials.from_service_account_file(google_credentials)

# sistema comando api urls
comando_api = 'https://api.dados.rio/v2/adm_cor_comando'
comando_api_v1 = 'https://api.dados.rio/v1/cor/comando'
pops_url = comando_api + '/pops'
pop_url = comando_api + '/procedimento_operacional_padrao_orgaos_responsaveis'
events_url = comando_api + '/ocorrencias'
events_url_v1 = comando_api_v1 + '/eventos'
open_events_url = comando_api + '/ocorrencias_abertas'
event_actions_url = comando_api + '/ocorrencias_orgaos_responsaveis'

# local resources urls 
pops_path = '../../../../Apps/Python/bolsao-api/static/comando/pops.csv'
polygons_geojson_path = 'static/clusters/polygons_micro.geojson'

# Load POPs dataframe
pops_df = pd.read_csv(pops_path)
pops_map = pops_df.set_index('id')['titulo'].to_dict()

# ---
# Sistema Comando's API and BigQuery database wrapper

def comando_open_events():
    open_events_res = {'error': ''}
    attempts, max_retries = 0, 20
    while ('eventos' not in list(open_events_res.keys())) and (attempts <= max_retries):
        attempts += 1
        if attempts != 1: print(f'Comando open events: Request attempt ({attempts}) ' + (open_events_res['error'] if 'error' in open_events_res.keys() else 'Error'))
        open_events_res = requests.get(open_events_url).json()
    open_events_df =  pd.DataFrame(open_events_res['eventos'])
    open_events_df['pop_titulo'] = open_events_df['pop_id'].map(pops_map) # include event pop title
    return open_events_df.fillna('')

def dict_df_map(obj):
    if 'eventos' in obj.keys():
        return pd.DataFrame((obj['eventos']))
    else:
        return pd.DataFrame([obj])

def build_rqst_params(start, now, offset, url):
    n_requests = int(np.ceil((now - start) / (offset)))
    rqst_params = []
    for n in range(n_requests):
        events_res = {}
        if n != 0: start = start + offset
        end = start + offset
        start_str = start.isoformat().replace('T', ' ') + '.0' # replaca with str(start)
        end_str = end.isoformat().replace('T', ' ') + '.0'
        rqst_params.append({
            'start': start_str,
            'end': end_str,
            'url': url + f'?inicio={start_str}&fim={end_str}',
        })
    return pd.DataFrame(rqst_params)

def get_comando_api_history(
    start=datetime(2023,1,1), end=datetime.now(),
    offset=30*pd.offsets.Day(), n_attempts=10, version='v2',
    as_datetime=False,
):
    url = events_url if version == 'v2' else events_url_v1
    rqst_params = build_rqst_params(start, end, offset, url)
    n_requests = int(np.ceil((end - start) / offset))
    events_df, events_res = [], {}
    missing = False
    for idx, params in rqst_params.iterrows():
        j = 0; rqst_url = params['url']
        while 'eventos' not in events_res.keys():
            j+=1; print('Periods requested:', idx + 1, '/', n_requests,'\nCurrent retries:', str(j), '/', n_attempts)
            events_res = requests.get(rqst_url).json()
            if j >= n_attempts:
                missing = True; break
        events_df.append(pd.concat(list(map(dict_df_map, events_res['eventos']))))
        events_res = {}
    events_df = pd.concat(events_df).reset_index(drop=True)
    events_df['pop_titulo'] = events_df['pop_id'].map(pops_map) # include event pop title
    if as_datetime:
        events_df['inicio'] = events_df['inicio'].astype('datetime64[ms]')
        events_df['fim'] = events_df['fim'].astype('datetime64[ms]')
    status = 'Success.' if not missing else 'Failed! Records missing.'
    print(f'Done! Status: {status}')
    return events_df.drop_duplicates(subset='id')

def get_comando_api_open_events(at=datetime.now(), between=7*pd.offsets.Day(), as_datetime=False, n_attempts=10):
    start, end = at - between, at + between
    events = get_comando_api_history(
        start, end, n_attempts=n_attempts,
        as_datetime=True,
    )
    events = events[(events['inicio'] <= at) & (events['fim'] >= at)]
    if not as_datetime:
        events['inicio'] = str(events['inicio'])
        events['fim'] = (events['fim'])
    
def bigquery_request(query):
    client = bigquery.Client(credentials=google_credentials)
    query_job = client.query(query)
    return pd.DataFrame(list(map(dict, query_job.result())))

def where_clause(items):
    return ' AND '.join([' '.join([key, 'IN', f"""({', '.join([f'"{value}"' for value in (values if type(values) is list else values.split(','))])})"""]) for key, values in items.items()])

def get_comando_bq_history(
    startdate=datetime(2023,1,1), enddate='now',
    where={}, select='*', as_datetime=False,
):
    comando_query = '''
    SELECT {} FROM `datario.adm_cor_comando.ocorrencias` WHERE data_particao BETWEEN "{}" AND "{}" AND data_inicio BETWEEN "{}" AND "{}"
    '''
    if enddate == 'now': enddate = datetime.now()
    if type(select) is list: select = ', '.join(select)
    query = comando_query.format(select, str(startdate.date()), str(enddate.date()), str(startdate), str(enddate))
    if len(where):
        query = ' '.join([query, 'AND', where_clause(where)])
    history = bigquery_request(query)
    if len(history):
        history['pop_titulo'] = history['id_pop'].astype(int).map(pops_map) # include event pop title
        if not as_datetime:
            history['data_particao'] = history['data_particao'].astype('str')
            history['data_inicio'] = history['data_inicio'].astype('str')
            history['data_fim'] = history['data_fim'].astype('str')
    return history

def get_comando_bq_open_events(
    at=datetime.now(), where={}, select='*',
    between=7*pd.offsets.Day(), as_datetime=False,
):
    comando_query = '''
    SELECT {} FROM `datario.adm_cor_comando.ocorrencias` WHERE data_particao BETWEEN "{}" AND "{}" AND data_inicio <= "{}" AND data_fim >= "{}"
    '''
    start, end = at - between / 2, at + between / 2
    query = comando_query.format(select, str(start.date()), str(end.date()), str(at), str(at))
    if len(where):
        query = ' '.join([query, 'AND', where_clause(where)])
    history = bigquery_request(query)
    if len(history):
        history['pop_titulo'] = history['id_pop'].astype(int).map(pops_map) # include event pop title
        if not as_datetime:
            history['data_particao'] = history['data_particao'].astype('str')
            history['data_inicio'] = history['data_inicio'].astype(str)
            history['data_fim'] = history['data_fim'].astype(str)
    return history
```
