```python
from urllib.parse import urlencode

# Python modules
import requests, json, urllib
import pandas as pd, numpy as np
from matplotlib.path import Path as mpl_path

# Sistema comando api urls
api_root = 'https://api.dados.rio/v2/adm_cor_comando'
pops_url = api_root + '/pops'
pop_url = api_root + '/procedimento_operacional_padrao_orgaos_responsaveis'
events_url = api_root + '/ocorrencias'
open_events_url = api_root + '/ocorrencias_abertas'
event_actions_url = api_root + '/ocorrencias_orgaos_responsaveis'

# local resources urls 
pops_path = 'static/comando/pops.csv'
polygons_geojson_path = 'static/clusters/polygons_micro.geojson'

# Load POPs dataframe
pops_df = pd.read_csv(pops_path)
pops_map = pops_df.set_index('id')['titulo'].to_dict()

# ---
# Sistema Comando API - Clusters Polygons Extension Methods

def comando_events():
    open_events_res = {'error': ''}
    attempts, max_retries = 0, 20
    while ('eventos' not in list(open_events_res.keys())) and (attempts <= max_retries):
        attempts += 1
        if attempts != 1: print(f'Comando open events: Request attempt ({attempts}) ' + (open_events_res['error'] if 'error' in open_events_res.keys() else 'Error'))
        open_events_res = requests.get(open_events_url).json()
    open_events_df =  pd.DataFrame(open_events_res['eventos'])
    open_events_df['pop_titulo'] = open_events_df['pop_id'].map(pops_map) # include event pop title
    return open_events_df.fillna('')
```
