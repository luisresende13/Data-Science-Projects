# Apresentação ao comitê técnico

### 1. Introdução (Por que modelos preditivos de bolsões?)
    
1. Desafio proposto

1. Contexto e motivação: Ponto de vista da cidade e do coordenador de cidades do COR-Rio (Marcelo).
    1. É tarefa do marcelo monitorar regiões suscetíveis e vulneráveis durante chuvas intensas. Para isso o marcelo já conta com recursos e ferramentas como:
    1. Acesso à leitura 
    1. Porém, não é óbvio inferir se haverá de fato bolsão em uma região.  baseado somente pela intensidade da chuva e sua intuição.
    1. É trabalhoso gerênciar os múltiplos pontos de alta recorrência no momento da chuva interna.
    1. Para o Marcelo, coordenador de cidade do COR-Rio, 30 minutos de antecedência no tempo de resposta de antecedência na identificação de um bolsão faria toda a diferença para p gerenciamento das operações desencadeadas  (verificação, envio de pessoal, fechamento de vias, descentupimento de bueiros, manutenção da drenagem).

1. Produtos em mvp (dois meses de aceleração)
    1. Catalogo de bolsões expandido: Sistema Comando + Geocodificação Google (Endereços)
    1. Agrupamento aumático e descrição de áreas com alta recorrência de bolsão.
    1. Modelos preditivos de bolsões em geral (em api ou em aplicação).
    1. Aplicação exclusiva para o *Gerenciamento*, *Monitoramento* *Previsão* e *Alerta* de bolsões na cidade.

1. Outros produtos/resultados:
    1. Avaliação de Dados: Sistema Comando + Históricos Meteorógicos
    1. Pipelines de transformação do dados: Utilização dos dados em tempo real para diversas finalidades.
    1. Metodologia e datasets de modelagem estabelecidos e testados.

1. O que são os modelos?
    1. Modelos meteorológicos para previsão da formação de bolsão/acúmulo-de-água em pontos críticos da cidade com alta recorrência.
    1. Uso de inteligência artificial com analise de grande quantidade de dados para gerar previsões com alta confiança.
    1. Pensados como funcionalidade para uma aplicação:
    1. Possibilidade de múltiplos modelos (Falar isso depois da analise de cluster)
        1. Monitoramento macro:
            1. Médias mensal e semanal da cidade e regiões macro: (**ADICIONAR AO DASHBAORD** !!!!!)
                1. Bem definidas.
                1. Forecast de série temporal
                1. De forma simples é o número de bolsões esperado (médio) e a probabilidade por semana/mês, para a cidade e regiões macro.
            1. Probabilidade e alerta : Cidade e regiões. (Testar precisão modelo de cidade)
        1. Monitoramento micro:
            1. Probabilidade e alerta : Pontos de alta recorrência (Catete, Armando Lombardi...)

1. Como funcionam os modelos (para pontos de alta recorrência):
    1. O modelo conhece como estavam as condições meteorológicas (das estações) antes e durante a ocorrência de bolsões nas diferentes localidades.
    1. Nesse contexto as condições meteorológicas são os conjuntos dos dados gerados pelas estações em um determinado ponto no tempo.
    1. Probabilidade baseada na "configuração meteorológica" atual das estações, e no aprendizado com o histórico.
    1. Dessa maneira, após treinado, a partir das leituras mais atuais das estações ele é capaz de discernir se há probabilidade de bolsão para a próxima hora.

    2. Intervalo de previsão de 15 em 15 minutos (frequência das estações Alerta-Rio.)
    1. Alcance de previsão fixo de 1h (Referência, com variação de até três horas - medir essa variação)
    1. Exemplos:
        1. Se não há chuva: probabilidade zero.
        1. Se há chuva em um ponto "isolado", distante ou de baixa recorrência de bolsão: probabilidade baixa.
        1. Se há chuva moderada em uma região de recorrência: probabilidade média 
        1. Se há chuva forte em região recorrente, probabailide alta.
