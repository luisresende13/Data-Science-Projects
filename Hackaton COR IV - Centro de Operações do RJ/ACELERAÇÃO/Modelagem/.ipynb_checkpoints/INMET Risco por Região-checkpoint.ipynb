# Preliminary modeling - Hourly incident model per group

#### Import preprocessing and machine learning modules

```python
from sklearn.preprocessing import MinMaxScaler as mms
from Modulos.cv_samplers import GroupUnderSampleSplit, print_cls_cnt

from sklearn.utils import all_estimators
from sklearn.metrics import classification_report as cr

classifiers = dict(all_estimators('classifier'))
gbc = classifiers['GradientBoostingClassifier']
```

#### Time series feature dataset transformatio pipeline

```python
def TimeseriesTransformPipeline(data, min_time=None, cut=None, drop_empty_cols=False, scale=False, interpolate='linear', fillna='min'):
    
    X = data.copy()
    print('Initial data:', X.shape)
    if min_time is not None:
        #### Extract features dataset - Time serie section after first recorded incident
        X = X[data.index > min_time]
        print('Time extraction:', X.shape)
    
    if cut is not None:
        X = X.iloc[:cut].copy()

    if drop_empty_cols:
        #### Drop X empty columns and rows
        X.dropna(1, how='all', inplace=True)
        print('Drop empty columns: ', X.shape)

    if scale:
        #### Scale X
        X[X.columns] = mms().fit_transform(X)

    if interpolate is not None:
        #### Interpolate X missing values
        X = X.interpolate(interpolate)

    if fillna is not None:
        #### Fill missing values with the minimum column value
        for col in X:
            if fillna=='min':
                fill_value = X[col].min()
            else:
                fill_value = 0
            X[col].fillna(fill_value, inplace=True)
    
    return X
```

#### Binary classification pipeline functions

```python
class ClassificationPipeline:
    
    def __init__(
        self, # X, Y, model=gbc(),
        train_size=0.79, train_prct=1,
        test_size=0.2, test_prct=None,
        shuffle=True, random_state=None,
    ):
        self.train_size=train_size; self.test_size=test_size;
        self.train_prct=train_prct; self.test_prct=test_prct;
        self.shuffle=shuffle; self.random_state=random_state

    # Binary classification pipeline function
    def binary(
        self, X, Y, model=gbc(),
        return_cls_cnt=False
    ):
        #### Splitter Instance
        guss = GroupUnderSampleSplit(
            train_size=self.train_size, train_prct=self.train_prct,
            test_size=self.test_size, test_prct=self.test_prct,
        )
        #### Train and test x and y dataframes
        xt, xe, yt, ye = guss.undersample(
            X, Y, shuffle=self.shuffle, random_state=self.random_state
        )
        cls_cnt = print_cls_cnt(Y, yt.index, ye.index, display_cnt=(not return_cls_cnt), return_cnt=return_cls_cnt)
        #### Fit model with train data
        model.fit(xt, yt)
        #### Make predictions
        yhat = model.predict(xe)
        #### Evaluate model prediction
        scrs = pd.DataFrame(cr(ye, yhat, output_dict=True)).T
        if return_cls_cnt:
            return scrs, cls_cnt
        return scrs

    # Multiple targets binary classification pipeline function - Changing Traget variable
    def binary_multi_target(
        self,
        X, Yi, model=gbc(),
    ):
        scrs, cls_cnts = {}, {}
        for i, group in enumerate(Yi):
            co(wait=True); print(f'{i+1}/{Yi.shape[1]} models evaluated.')
            scrs[group], cls_cnts[group] = self.binary(
                X, Yi[group], model=model,
                return_cls_cnt=True
            )
        scrs, cls_cnts = pd.concat(scrs), pd.concat(cls_cnts)
        scrs.index.names = ['group', 'metric']
        cls_cnts.index.names = ['Group', 'Class']
        return scrs, cls_cnts
```

## Data transformation

#### Predictive Variables

```python
event_min_time = events['EVENTO_INICIO'].min()

X = TimeseriesTransformPipeline(
    inmet, event_min_time, cut=-1,
    drop_empty_cols=True, scale=True,
    interpolate='linear', fillna='min'
)
```

#### Target variable

```python
Y = group_ts[0].loc[X.index].copy()  # Extract target label variable for random group
```

### Instatiate Classification Pipeline Object

```python
CP = ClassificationPipeline(
    train_size=0.79, train_prct=1,
    test_size=0.2, test_prct=None,
    shuffle=True, random_state=0,
)
```

## Evaluate single event group

```python
score, cls_cnt = CP.binary(
    X, Y, model=gbc(), return_cls_cnt=True
)

display(score, cls_cnt)
```

## Evaluate multiple event groups

```python
scrs, cls_cnts = CP.binary_multi_target(
    X, Yi=group_ts.loc[X.index].iloc[:, :], model=gbc(),
#     train_size=0.79, train_prct=1,
#     test_size=0.2, test_prct=None,
#     shuffle=True, random_state=None,
)
```

### Result analysis

#### Group recall per class - Sorted

```python
def classesGroupRecall(scrs, cls_cnts, ignore_first=5):

    fig, ax = plt.subplots(1, 3, figsize=(20, 3.5))

    scrs['recall'].loc[:, ['0', '1', 'macro avg']].unstack('metric').sort_values('1').plot(use_index=False, ax=ax[0])
    ax[0].set(
        title='Groups Recall per Class - Minority class sorted',
        xlabel='Group',
        ylabel='Recall'
    )

    sort_scrs = scrs['recall'].loc[:, ['0', '1', 'macro avg']].unstack('metric').sort_values('macro avg')
    sort_scrs.plot(use_index=False, ax=ax[1])
    ax[1].set(
        title='Groups Recall per Class - Macro avg sorted',
        xlabel='Group',
        ylabel='Recall'
    )



    train_size, test_size = (mms().fit_transform(cls_cnts[col].loc[sort_scrs.index[ignore_first:], 1].to_frame(col)).reshape(-1).tolist() for col in ['Train set', 'Test set'])
    train_size = [np.nan]*ignore_first + train_size 
    test_size = [np.nan]*ignore_first + test_size 

    ax[2].plot(train_size)
    ax[2].plot(test_size)

    return ax
```

```python
axs = classesGroupRecall(scrs, cls_cnts, ignore_first=5)
```

# Filter hour records by variables threshold

#### Rows (records) left after filterging by column minimum threshold (for each threshold in list provided)

```python
def multi_min_msk(data, desc, stat='min', acumulate=None, top_down=True):

    filter_stats = []
    for col in data:
        min_value = desc.loc[stat, col]
        above_min = data[col].fillna(min_value) >= min_value
        filter_stats.append([col, min_value, above_min.sum(), above_min.sum()/len(data)*100])
    filter_stats = pd.DataFrame(filter_stats, columns=['column', 'threshold', 'rows left', 'rows left (%)'])
    filter_stats = filter_stats.sort_values('rows left', ascending=top_down).reset_index(drop=True)

    if acumulate is not None:
        cum_stats = []
        for col in filter_stats['column']:
            min_value = desc.loc[stat, col]
            above_min = data[col].fillna(min_value) >= min_value
            if col == filter_stats['column'].iloc[0]:
                above_min_cum = above_min
            else:
                if acumulate=='or':
                    above_min_cum = above_min_cum | above_min
                elif acumulate=='and':
                    above_min_cum = above_min_cum & above_min
            cum_stats.append([col, min_value, above_min_cum.sum(), above_min_cum.sum()/len(data)*100])
        cum_stats = pd.DataFrame(cum_stats, columns=['column', 'threshold', 'rows left cum', 'rows left cum (%)'])
        return filter_stats, cum_stats
    
    return filter_stats
```

# Evaluate filtered Data

```python
event_min_time = events['EVENTO_INICIO'].min()

X = TimeseriesTransformPipeline(
    inmet, event_min_time, cut=-1,
    drop_empty_cols=True, scale=True,
    interpolate=None, fillna=None
)
```

#### Example event group

```python
Y_ex = group_ts[top_grp].loc[X.index]

display(Y_ex.value_counts().to_frame('Class Count'))
```

#### Extract minority classe records and calculate variables statistics

```python
mino_msk = Y_ex == 1
mino = X[mino_msk]
mino_desc = mino.describe()

stats, cum_stats = multi_min_msk(X, mino_desc, stat='min', acumulate='and', top_down=True)
```

### Filter records

```python
n_filters = len(stats)
margin = 0.1

X_filt = X.copy() 
print(X_filt.shape)

for col in stats['column'].iloc[:n_filters]:
    min_value = mino_desc.loc['min', col]
    threshold =  min_value + min_value * margin
    X_filt = X_filt[X_filt[col].fillna(threshold) >= threshold]

print(X_filt.shape); print(round(X_filt.shape[0] / X.shape[0] * 100, 1), '%')
```

### Transform filtered data

```python
X_filt = TimeseriesTransformPipeline(
    X_filt, interpolate='linear', fillna='min'
)
```

### Filter target variable

```python
Y_filt = Y_ex.loc[X_filt.index]
display(Y_filt.value_counts().to_frame('Class Count'))
```

### Instatiate Classification Pipeline Object

```python
CP = ClassificationPipeline(
    train_size=0.79, train_prct=1,
    test_size=0.2, test_prct=None,
    shuffle=True, random_state=0,
)
```

## Evaluate single event group

```python
score, cls_cnt = CP.binary(
    X_filt, Y_filt, model=gbc(), return_cls_cnt=True
)

display(score, cls_cnt)
```

```python
scrs, cls_cnts = CP.binary_multi_target(
    X_filt, Yi=group_ts.loc[X_filt.index], model=gbc(),
#     train_size=0.79, train_prct=1,
#     test_size=0.2, test_prct=None,
#     shuffle=True, random_state=None,
)
```
