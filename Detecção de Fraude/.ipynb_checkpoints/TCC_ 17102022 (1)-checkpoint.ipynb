```python
def gerador_perfil_cliente_table(n_clientes, random_state=0):
    
    np.random.seed(random_state)
        
    propriedade_cliente=[]
    
    # Generate customer properties from random distributions 
    for cliente_ID in range(n_clientes):
        
        x_cliente_ID = np.random.uniform(0,100)
        y_cliente_ID = np.random.uniform(0,100)
        
        media_valor = np.random.uniform(5,100) # Arbitrary (but sensible) value 
        dp_valor = media_valor/2 # Arbitrary (but sensible) value
        
        media_trans_pdia = np.random.uniform(0,4) # Arbitrary (but sensible) value 
        
        propriedade_cliente.append([cliente_ID,
                                      x_cliente_ID, y_cliente_ID,
                                      media_valor, dp_valor,
                                      media_trans_pdia])
        
    perfil_cliente_table = pd.DataFrame(propriedade_cliente, columns=['cliente_ID',
                                                                      'x_cliente_ID', 'y_cliente_ID',
                                                                      'media_valor', 'dp_valor',
                                                                      'media_trans_pdia'])
    
    return perfil_cliente_table
```

```python
n_cliente = 5
perfil_cliente_table = gerador_perfil_cliente_table(n_cliente, random_state = 0)
perfil_cliente_table
```

```python
def gerador_perfil_terminal_table(n_terminals, random_state=0):
    
    np.random.seed(random_state)
        
    terminal_id_properties=[]
    
    # Gere propriedades de terminal a partir de distribuições aleatórias
    for terminal_ID in range(n_terminals):
        
        x_terminal_id = np.random.uniform(0,100)
        y_terminal_id = np.random.uniform(0,100)
        
        terminal_id_properties.append([terminal_ID,
                                      x_terminal_id, y_terminal_id])
                                       
    perfil_terminal_table = pd.DataFrame(terminal_id_properties, columns=['terminal_ID',
                                                                      'x_terminal_id', 'y_terminal_id'])
    
    return perfil_terminal_table
```

```python
n_terminais = 5
perfil_terminal_table = gerador_perfil_terminal_table(n_terminais, random_state = 0)
perfil_terminal_table
```

```python
def get_list_terminais_dentro_raio(perfil_cliente, x_y_terminais, r):
    
    # Use matrizes numpy a seguir para acelerar os cálculos
    
    # Localização (x,y) do cliente como matriz numpy
    x_y_cliente = perfil_cliente[['x_cliente_ID','y_cliente_ID']].values.astype(float)
    
    # Diferença quadrada nas coordenadas entre os locais do cliente e do terminal
    squared_diff_x_y = np.square(x_y_cliente - x_y_terminais)
    
    # Somar ao longo das linhas e calcular a raiz suared para obter a distância
    dist_x_y = np.sqrt(np.sum(squared_diff_x_y, axis=1))
    
    # Obtenha os índices dos terminais que estão a uma distância menor que r
    terminal_disponivel = list(np.where(dist_x_y<r)[0])
    
    # Retornar a lista de IDs de terminal
    return terminal_disponivel
```

```python

def gerador_transacoes_table(perfil_cliente, start_date = "2022-01-01", nb_dias = 10):
    
    clientes_transacoes = []
    
    random.seed(int(perfil_cliente.cliente_ID))
    np.random.seed(int(perfil_cliente.cliente_ID))
    
    # Para todos os dias
    for dia in range(nb_dias):
        
        # Número aleatório de transações para esse dia
        nb_TS = np.random.poisson(perfil_cliente.media_trans_pdia)
        
        # Se nb_TS for positivo, vamos gerar transações
        if nb_TS>0:
            
            for TS in range(nb_TS):
                
    
                # Hora da transação: Por volta do meio-dia, std 20.000 segundos. Essa escolha visa simular o fato de que
                # a maioria das transações ocorre durante o dia.
                time_TS = int(np.random.normal(86400/2, 20000))
                
                # Se o tempo de transação estiver entre 0 e 86400, vamos mantê-lo, caso contrário, vamos descartá-lo
                if (time_TS>0) and (time_TS<86400):
                    
                    # valor is drawn from a normal distribution  
                    valor = np.random.normal(perfil_cliente.media_valor, perfil_cliente.dp_valor)
                    
                   
                    # Se valor negativo, extraia de uma distribuição uniforme
                    if valor<0:
                        valor = np.random.uniform(0,perfil_cliente.media_valor*2)
                    
                    valor=np.round(valor,decimals=2)
                    
                    if len(perfil_cliente.terminal_disponivel)>0:
                        
                        terminal_id = random.choice(perfil_cliente.terminal_disponivel)
                    
                        clientes_transacoes.append([time_TS+dia*86400, dia,
                                                      perfil_cliente.cliente_ID, 
                                                      terminal_id, valor])
            
    clientes_transacoes = pd.DataFrame(clientes_transacoes, columns=['TS_segundos', 'TS_dias', 'cliente_ID', 'terminal_ID', 'TS_valor'])
    
    if len(clientes_transacoes)>0:
        clientes_transacoes['TS_data'] = pd.to_datetime(clientes_transacoes["TS_segundos"], unit='s', origin=start_date)
        clientes_transacoes=clientes_transacoes[['TS_data','cliente_ID', 'terminal_ID', 'TS_valor','TS_segundos', 'TS_dias']]
    
    return clientes_transacoes  
```

```python
def generate_dataset(n_clientes = 10000, n_terminals = 1000000, nb_dias=90, start_date="2018-04-01", r=5):
    
    start_time=time.time()
    perfil_cliente_table = gerador_perfil_cliente_table(n_clientes, random_state = 0)
    print("Tempo para gerar a tabela de perfis de clientes: {0:.2}s".format(time.time()-start_time))
    
    start_time=time.time()
    perfil_terminal_table = gerador_perfil_terminal_table(n_terminals, random_state = 1)
    print("Tempo para gerar a tabela de perfis de terminal: {0:.2}s".format(time.time()-start_time))
    
    start_time=time.time()
    x_y_terminais = perfil_terminal_table[['x_terminal_id','y_terminal_id']].values.astype(float)
    perfil_cliente_table['terminal_disponivel'] = perfil_cliente_table.apply(lambda x : get_list_terminais_dentro_raio(x, x_y_terminais=x_y_terminais, r=r), axis=1)
    # With Pandarallel
    #perfil_cliente_tableerminal_disponivels'] = perfil_cliente_table.parallel_apply(lambda x : get_list_closest_terminals(x, x_y_terminals=x_y_terminals, r=r), axis=1)
    perfil_cliente_table['nb_terminals']=perfil_cliente_table.terminal_disponivel.apply(len)
    print("Tempo para associar terminais aos clientes: {0:.2}s".format(time.time()-start_time))
    
    start_time=time.time()
    transacoes_df=perfil_cliente_table.groupby('cliente_ID').apply(lambda x : gerador_transacoes_table(x.iloc[0], nb_dias=nb_dias)).reset_index(drop=True)
    # With Pandarallel
    #transacoes_df=perfil_cliente_table.groupby('cliente_ID').parallel_apply(lambda x : generate_transactions_table(x.iloc[0], nb_dias=nb_dias)).reset_index(drop=True)
    print("Tempo para gerar transações: {0:.2}s".format(time.time()-start_time))
    
    # Classifique as transações cronologicamente
    transacoes_df=transacoes_df.sort_values('TS_data')
    # Redefinir índices, a partir de 0
    transacoes_df.reset_index(inplace=True,drop=True)
    transacoes_df.reset_index(inplace=True)
    # transacao_ID são os índices de dataframe, começando em 0
    transacoes_df.rename(columns = {'index':'transacao_ID'}, inplace = True)
    
    return (perfil_cliente_table, perfil_terminal_table, transacoes_df)
```

```python
(perfil_clientes_table, perfil_terminal_table, transacoes_df)=\
    generate_dataset(n_clientes = 5000, 
                     n_terminals = 10000, 
                     nb_dias=183, 
                     start_date="2022-01-01", 
                     r=5)
```

```python
def add_fraudes(perfil_cliente_table, terminal_profiles_table, transacoes_df):
    
    # Por padrão, todas as transações são genuínas
    transacoes_df['TS_Fraude']=0
    transacoes_df['TS_Cenario_fraude']=0
    
    # Cenário 1
    transacoes_df.loc[transacoes_df.TS_valor>220, 'TS_Fraude']=1
    transacoes_df.loc[transacoes_df.TS_valor>220, 'TS_Cenario_fraude']=1
    NB_FRAUDE_cenario_1=transacoes_df.TS_Fraude.sum()
    print("Número de fraudes do cenário 1: "+str(NB_FRAUDE_cenario_1))
    
    # Cenário 2
    for dia in range(transacoes_df.TS_dias.max()):
        
        compromet_terminais = perfil_terminal_table.terminal_ID.sample(n=2, random_state=dia)
        
        compromised_transactions=transacoes_df[(transacoes_df.TS_dias>=dia) & 
                                                    (transacoes_df.TS_dias<dia+28) & 
                                                    (transacoes_df.terminal_ID.isin(compromet_terminais))]
                            
        transacoes_df.loc[compromised_transactions.index,'TS_Fraude']=1
        transacoes_df.loc[compromised_transactions.index,'TS_Cenario_fraude']=2
    
    NB_FRAUDE_cenario_2=transacoes_df.TS_Fraude.sum()-NB_FRAUDE_cenario_1
    print("Número de fraudes do cenário 2: "+str(NB_FRAUDE_cenario_2))
    
    # Cenário 3
    for dia in range(transacoes_df.TS_dias.max()):
        
        compromised_customers = perfil_cliente_table.cliente_ID.sample(n=3, random_state=dia).values
        
        compromised_transactions=transacoes_df[(transacoes_df.TS_dias>=dia) & 
                                                    (transacoes_df.TS_dias<dia+14) & 
                                                    (transacoes_df.cliente_ID.isin(compromised_customers))]
        
        nb_compromised_transactions=len(compromised_transactions)
        
        
        random.seed(dia)
        index_fauds = random.sample(list(compromised_transactions.index.values),k=int(nb_compromised_transactions/3))
        
        transacoes_df.loc[index_fauds,'TS_valor']=transacoes_df.loc[index_fauds,'TS_valor']*5
        transacoes_df.loc[index_fauds,'TS_Fraude']=1
        transacoes_df.loc[index_fauds,'TS_Cenario_fraude']=3
        
                             
    NB_FRAUDE_cenario_3=transacoes_df.TS_Fraude.sum()-NB_FRAUDE_cenario_2-NB_FRAUDE_cenario_1
    print("Número de fraudes do cenário 3: "+str(NB_FRAUDE_cenario_3))
    
    return transacoes_df     
```

```python
def get_staTS(transacoes_df):
    #Número de transações por dia
    nb_TS_per_dia=transacoes_df.groupby(['TS_dias'])['cliente_ID'].count()
    #Número de transações fraudulentas por dia
    NB_FRAUDE_per_dia=transacoes_df.groupby(['TS_dias'])['TS_Fraude'].sum()
    #Número de cartões fraudulentos por dia
    NB_FRAUDEcard_per_dia=transacoes_df[transacoes_df['TS_Fraude']>0].groupby(['TS_dias']).cliente_ID.nunique()
    
    return (nb_TS_per_dia,NB_FRAUDE_per_dia,NB_FRAUDEcard_per_dia)

(nb_TS_per_dia,NB_FRAUDE_per_dia,NB_FRAUDEcard_per_dia)=get_staTS(transacoes_df)

n_dias=len(nb_TS_per_dia)
TS_stats=pd.DataFrame({"value":pd.concat([nb_TS_per_dia/50,NB_FRAUDE_per_dia,NB_FRAUDEcard_per_dia])})
TS_stats['stat_type']=["nb_TS_per_dia"]*n_dias+["NB_FRAUDE_per_dia"]*n_dias+["NB_FRAUDEcard_per_dia"]*n_dias
TS_stats=TS_stats.reset_index()
```

```python
def Fim_Dsemana(TS_data):
# Transforma a data em dia da semana (0 é segunda-feira, 6 é domingo)
    Dia_semana = TS_data.weekday()
#Valor binário: 0 se dia da semana, 1 se fim de semana
    Fim_Dsemana = Dia_semana>=5        
    return int(Fim_Dsemana)
```

```python
def DNoite(TS_data):
    # Get the hour of the transaction
    TS_hora = TS_data.hour
    # Binary value: 1 if hour less than 6, and 0 otherwise
    DNoite = TS_hora<=6      
    return int(DNoite)
```

```python
def get_customer_spending_behaviour_features(customer_transactions, windows_size_in_days=[1,7,30]):
    # Vamos fazer transações de primeira ordem cronologicamente
    customer_transactions=customer_transactions.sort_values('TS_data')
    # A data e a hora da transação são definidas como o índice, o que permitirá o uso da função de rolagem
    customer_transactions.index=customer_transactions.TS_data
    # Para cada tamanho de janela
    for window_size in windows_size_in_days:
        # Calcula a soma dos valores das transações e o número de transações para o tamanho de janela fornecido
        SUM_VALOR_TS_JANELA=customer_transactions['TS_valor'].rolling(str(window_size)+'d').sum()
        NB_TS_JANELA=customer_transactions['TS_valor'].rolling(str(window_size)+'d').count()
        # Calcula o valor médio da transação para o tamanho de janela fornecido   
        # NB_TS_JANELA é sempre >0, pois a transação atual é sempre incluída
        MEDIA_VALOR_TS_JANELA=SUM_VALOR_TS_JANELA/NB_TS_JANELA  
        # Salva os valores dos recursos
        customer_transactions['cliente_ID_NB_TS_'+str(window_size)+'dia_janela']=list(NB_TS_JANELA)
        customer_transactions['cliente_ID_MEDIA_VALOR_'+str(window_size)+'dia_janela']=list(MEDIA_VALOR_TS_JANELA)
    # Reindexar de acordo com os IDs das transações
    customer_transactions.index=customer_transactions.transacao_ID
    
    # E retorne o dataframe com os novos recursos
    return customer_transactions
```

```python
def get_count_risk_rolling_window(terminal_transacoes, periodo_atrasado=7, windows_size_in_days=[1,7,30], feature="terminal_ID"):
    terminal_transacoes=terminal_transacoes.sort_values('TS_data')
    terminal_transacoes.index=terminal_transacoes.TS_data
    NB_FRAUDE_DELAY=terminal_transacoes['TS_Fraude'].rolling(str(periodo_atrasado)+'d').sum()
    NB_TS_DELAY=terminal_transacoes['TS_Fraude'].rolling(str(periodo_atrasado)+'d').count()
    for window_size in windows_size_in_days:
        NB_FRAUDE_DELAY_WINDOW=terminal_transacoes['TS_Fraude'].rolling(str(periodo_atrasado+window_size)+'d').sum()
        NB_TS_DELAY_WINDOW=terminal_transacoes['TS_Fraude'].rolling(str(periodo_atrasado+window_size)+'d').count()
        NB_FRAUDE_WINDOW=NB_FRAUDE_DELAY_WINDOW-NB_FRAUDE_DELAY
        NB_TS_JANELA=NB_TS_DELAY_WINDOW-NB_TS_DELAY
        RISK_WINDOW=NB_FRAUDE_WINDOW/NB_TS_JANELA
        terminal_transacoes[feature+'_NB_TS_'+str(window_size)+'dia_janela']=list(NB_TS_JANELA)
        terminal_transacoes[feature+'_RISK_'+str(window_size)+'dia_janela']=list(RISK_WINDOW)
    terminal_transacoes.index=terminal_transacoes.transacao_ID
    # Replace NA values with 0 (all undefined risk scores where NB_TS_JANELA is 0) 
    terminal_transacoes.fillna(0,inplace=True)
    return terminal_transacoes
```
