# Extraindo Séries Temporais

#### Extração de séries temporais por cidade e produto

```python
ts = data.set_index(data['year'].astype('datetime64[D]')).drop('year', 1); ts.tail()
```

```python
unique_city_codes = cities = ts['city_code'].unique()
unique_products = products = ts['product'].unique()

ts_ij = {}
for city_code in unique_city_codes:
    
    city_ts = ts[ ts['city_code']==city_code ]
    ts_i = {}
    for product in unique_products:
        
        prod_ts = city_ts[ city_ts['product']==product ].sort_values('year').copy()
        if product=='Others':
            for prod_type in ['temporary', 'permanent']:
                prod_type_ts = prod_ts[ prod_ts['product_type']==prod_type ].copy()
                if len(prod_type_ts):
                    ts_i[ product+'-'+prod_type ] = prod_type_ts
        else:
            ts_i[product] = prod_ts
 
    ts_ij[city_code] = ts_i
```

### Contagem de valores vazios

```python
target='destinated_area'
```

```python
stats = []
for city in unique_city_codes:
    i = ts_ij[city]
    for product in [i for i in unique_products if i!='Others'] + ['Others-temporary', 'Others-permanent']:
        
        if product in i.keys():
            ij = i[product]
            n = len(ij)
            n_empty = ij[target].isna().sum()
            p_empty = n_empty / n
            stats.append([ city, product, n, n_empty, p_empty ])            
            
stats = pd.DataFrame(stats, columns=['city', 'product', 'n', 'n_empty', 'p_empty'])
```

### Extraindo área por cidade-produto-ano

```python
years = data['year'].unique()
```

```python
year_rows = {year: [] for year in years}
cnt=0
for city in cities:
    for product in prods:
        cnt+=1;
        if cnt in range(1, 1500, 40): co(wait=True); print(f'{cnt}/{len(cities)*len(prods)}')

        ts_i = ts_ij[city]
        if product in ts_i.keys():
            ts = ts_i[product]

            for year in years:
                try:
                    year_rows[year].append(ts[year[-4:]].values[0][-1])
                except:
                    year_rows[year].append(np.nan)
            
```

```python
col_names = [ city+'-'+product for city in cities for product in prods if product in ts_ij[city].keys() ]
year_area = pd.DataFrame(year_rows, index=col_names).fillna(0)
year_area = year_area.T.reset_index().rename(columns={'index': 'year'})
```
