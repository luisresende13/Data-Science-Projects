## Generating Random Dates Inside Date Interval

```python
def random_dates(start, end, n, unit='D'):

    ndays = (end - start).days + 1
    return pd.to_timedelta(np.random.rand(n) * ndays, unit=unit) + start

np.random.seed(None)
start = pd.to_datetime('2022-01-01')
end = pd.to_datetime('2022-01-18')
datetimes = random_dates(start, end, 100, unit='D').sort_values()

dates = pd.DatetimeIndex(np.array(datetimes, dtype='datetime64[D]'))
times = pd.DatetimeIndex(np.array(datetimes, dtype='datetime64[s]'))
times = [(str(i) if i<12 else str(i-12))+':'+str(j)+(' AM' if i<12 else ' PM') for i, j in zip(times.hour, times.minute)]
```

## Searching Reference Address Coordinates

```python
for i in range(20):

    print(f'Attempts: {i+1}/{20}'); clear_output(wait=True)
    try: 
        location = geolocator.geocode("Praça Raul Soares, Belo Horizonte")
        if not location:
            continue
        print('Localition Found!')
        break
    except:
        pass

location
```

## Setting Central Search Coords

```python
lat_home = location.latitude
lng_home = location.longitude
```

Latitude: 1 deg = 110574 Km
Longitude: 1 deg = 111.320*cos(latitude) km

## Conversion degree to Km

```python
lat_km_in_deg = 1/110574
lng_km_in_deg = lambda lat: 1/(111320*np.cos(lat*np.pi/360))
```

## Generating Random Coordinates

```python
np.random.seed(0)
search_range = 1500 # m

rand_lats = lat_home - lat_km_in_deg * search_range + np.random.random(100) * lat_km_in_deg * search_range * 2
rand_lngs = [lng_home - lng_km_in_deg(lat) * search_range + np.random.random() * lng_km_in_deg(lat) * search_range * 2 for lat in rand_lats]
rand_lngs_home = [lng_home - lng_km_in_deg(lat_home) * search_range + np.random.random() * lng_km_in_deg(lat_home) * search_range * 2 for lat in rand_lats]
rand_lngs = np.array(rand_lngs)
rand_lngs_home = np.array(rand_lngs_home)
lng_err = (rand_lngs-rand_lngs_home)/lng_km_in_deg(lat_home)

date_df = pd.DataFrame({'lat': rand_lats, 'lng': rand_lngs, 'lng_home': rand_lngs_home, 'err': lng_err})
```

```python
Map = folium.Map(location=(lat_home, lng_home), zoom_start=13.4, width='50%', height='50%')

for i in range(100):
    
    folium.Marker(location=(date_df.loc[i, 'lat'], date_df.loc[i, 'lng_home']), popup=str(i)).add_to(Map)

Map
```
