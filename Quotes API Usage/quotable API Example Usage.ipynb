### Make imports

```python
import requests, json, urllib, pandas as pd
```

```python
# api_url = r'https://zenquotes.io/api/'  # api base url
api_url = r'https://api.quotable.io/'
```

### Get random quote

```python
params = {
#     'key': api_key,
    'minLenght': 1,
    'maxLenght': 5000,
    'tags': 'famous-quotes|wisdom|friendship|inspirational|success',
#     'author': None,
#     'authorId': None,
}

endpoint = 'random?' # random, quotes, tags, authors, search/quotes, search/authors
url = api_url + endpoint # + urllib.parse.urlencode(params)
url
```

```python
payload, headers = {}, {}
response = requests.request("GET", url, headers=headers, data=payload, params=params)
res = json.loads(response.text); res
```

### Extra: Translate response

```python
import translators as ts

def translate(query):
    return ts.google(query_text=query, from_language='en', to_language='pt')

translate(res['content'])
```

```python
ts.google(query_text=res['content'], from_language='en', to_language='pt')
```

### Get authors

```python
endpoint = 'authors'
url = api_url + endpoint; print(url) # + urllib.parse.urlencode(params)
response = requests.request("GET", url, params={'limit': 1000, 'page': 1}) # ,'slug': author['slug']})
res = json.loads(response.text)
```
