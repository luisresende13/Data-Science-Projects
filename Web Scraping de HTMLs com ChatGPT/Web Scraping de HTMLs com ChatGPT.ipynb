### Parse multiple html documents from properties of geojson features · Assisted by with ChatGPT

```python
### Parse multiple html documents from properties of geojson features · Assisted by with ChatGPT

# This code takes a GeoJSON file containing HTML descriptions in the "description" property, 
# parses the HTML to extract table data, and adds the data as new properties to each feature.
# It then removes the "description" property and saves the updated GeoJSON file.

import json
from bs4 import BeautifulSoup

file_path = 'Hidrografia de Rios'

# Open the GeoJSON file
with open(f'{file_path}.geojson', 'r') as f:
    geojson_data = json.load(f)

# Loop through each feature
for feature in geojson_data['features']:
    # Parse the HTML code in the "description" property
    soup = BeautifulSoup(feature['properties']['description'], 'html.parser')
    table_rows = soup.find_all('tr')

    # Loop through each row in the table and add the data as new properties
    for row in table_rows:
        cells = row.find_all('td')
        if len(cells) >= 2:
            property_name = cells[0].get_text().strip()
            if len(property_name) < 100:
                # The property values might contain non-numeric characters such as commas
                # or periods, so we need to replace them with the appropriate decimal separator
                # for the locale. In this example, we're using the US locale, which uses periods
                # as the decimal separator.
                property_value = cells[1].get_text().strip().replace(',', '.')
                # Replace spaces in the property names with underscores
                property_name = property_name.replace(' ', '_')
                feature['properties'][property_name] = property_value
    
    # Remove the "description" property from the feature
    del feature['properties']['description']
    
    if "id" in feature:
        feature['properties']['ID'] = feature['id']
        del feature['id']

# Save the updated GeoJSON file
with open(f'{file_path} Fix.geojson', 'w') as f:
    json.dump(geojson_data, f)
    
```
